```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loaded the dataset from csv format and displayed top 5 values
E= pd.read_csv("/content/Employee Analysis Dataset.csv")
E.head()
```

```python
#renaming a column 
E.rename(columns={'Salary':'salary'}, inplace= True)
print(E)
```

```python
#using the groupby() function to combine 2 groups and mean() function to check the average

avg=E.groupby('Department')['salary'].mean()
print(avg)

#plotting a bar plot using seaborn syntax
sns.barplot(x='Department',y='salary',data=avg.reset_index())
```

```python
#using count function as we don't need an exact number, just enough value to check which department is highest
Department_employees = E.groupby('Department').count()

#idmax is used to return maximum value for a specific column 
Highest = Department_employees['salary'].idxmax()

print(Highest)
```

```python
#using value_counts to give us the exact number of values in a column
gender_counts = E['Gender'].value_counts()
print(gender_counts)

#making a seaborn plot
sns.set_style('whitegrid')
sns.countplot(x='Gender',data=E,palette="rainbow")
```

```python
#using the "corr" function here
correlation = E['salary'].corr(E['Years of Experience'])

#Checking for correlation
if correlation>0:
  print("There is a positive correlation among the 2 columns")
elif correlation<0:  
  print("There's a negative correlation among the 2 columns")
else:
  print('There is not correlation at all')

#plotting the same
plt.plot(E['Years of Experience'], E['salary'])
```

```python
Salaries = E.groupby('Department')['salary'].mean()

# sort the resulting dataframe by the mean salary in descending order
department_salaries_sorted = Salaries.sort_values(ascending=False)

# get the department with the highest average salary
highest_paid_department = department_salaries_sorted.index[0]

#making a bar plot for the same
department_salaries_sorted.plot.bar()

# set the title and labels for the plot
plt.title('Salary by Department')
plt.xlabel('Departments')
plt.ylabel('Salary scale')

plt.show()

print("The department with the highest average salary is:", highest_paid_department)
```
